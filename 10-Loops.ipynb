## for loop, while loop

# for loop

```python
my_list = [10,20,30,40,50,60,70]
```

```python
my_list[0] / 5 * 2
```

```python
my_list[1] / 5 * 2 #Yapmak yerine tek tek for loop kullanırsın
```

```python
for number in my_list: #benim listemdeki tüm numaralar için yazdık buraya aslında
    print(number / 5 * 2)
    #bu tab sırasında olduğu sürece hepsi for'a bağlı
```

```python
print("for loop started")
for x in my_list: #burada x dedik çünkü for'dan sonra değişken atıyoruz istediğini koy!!
    new_number = x / 5 * 2
    print(new_number)
print("for loop ended")
```

```python
#6'ya tam bölünebilenleri yazdırsın(30 ve 60 outcome almalı) for döngüsü ve if kontrolü
```

```python
for i in my_list: #bütün iler itemlar için benim listemdeki
    if i % 6 == 0: #6 mod == 0 yani item 6'ya bölündüğünde kalan 0 ise(true)
        print(f"Divided by 6: {i}") # 6'ya bölünenlerin içine bu itemı (sayıyı yaz)
    else: #aksi halde (diğerleri false olduğu için if bloğunda bunu yaptık direkt)
        print(f"Not divided by 6: {i}") #6'ya bölünemeyenlerin içine sayıyı yaz
```

## continue - break - pass

```python
my_list = [10,20,30,40,50,60,70]
```

```python
print("for loop started")
for num in my_list:
    print(num)
print("for loop ended")
```

```python
#döngülerde bazen bir şeyi yarıda kesmek isteyebiliriz mesela aradığımız bir rakam var
```

```python
#40 rakamını bulunca döngüden çık
```

```python
for i in my_list: #my_listteki her i(item) için bak
    print(i) #i'yi yazdır
    if i == 40: #eğer i 40'a eşitse
        print("yes")#yes yazdır
        break#40'a geldiğinde komutu kır durdur boz diyebiliriz buna
             #yani 40'a gelene kadar baktı kodda aşağıda 40'a kadar dedik aslında
```

```python
my_list
```

```python
my_new_list = [10,20,30,40,50,60,70]
```

```python
for i in my_new_list:
    if i == 40:
        continue #döngüye devam ediyoruz sonrasındaki koda bakmıyoruz
    print(i)
```

```python
my_list
```

```python
for i in my_list:
    pass #kodu geliştirirken düzeltmek için kodu yazabilirsin(pas geçsin diye)
# şimdi hata vermemesinin sebebi pass yazmış olmamız
```

## while loop

```python
# while (bu olmaya devam ettiği sürece bunu yap döngüsü)
```

```python
x = 0
```

```python
print("while loop started")
while x < 10:
    print(x)
    x += 1
print("while loop ended")
```

```python
last_list = [10,20,30,40,50]
```

```python
while 20 in last_list: #20 listede olduğu sürece
    print("20 in last list") #20 last listenin içinde yazdır
    last_list.pop()#popla ve yukarıda 20 listede olduğu sürece dediğimiz için 4 kere döner
```
