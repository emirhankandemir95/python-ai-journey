## inheritance

```python
class Musician(): #Müzisyen class'ı açtık
    def __init__(self,name):#Başlatıcıyı oluşturduk self koymak zorunlu zaten bir de name parametresi koyduk.
        self.name = name #self.name name'e eşitledik
        print("musician class") #ve bu class çalıştığında ya da birine atandığında bu yazsın dedik.
    
    def test1(self):#method1
        print("test1")#methodprint

    def test2(self):#method2
        print("test2")#methodprint2
```

```python
emir = Musician("Emir") #emir bir variable bir değişken bunu Musician("Emir") yazarak aslında hem class olarak ekledik
                        #Hem de yukarıda initializer'da isim tanımladık.
```

```python
emir.name #Name çağırdık ve "Emir" yazdı
```

```python
emir.test1()#Method1 çalışıyor
```

```python
emir.test2()#Method2'de çalışıyor
```

```python
class MusicianPlus(Musician):#Şimdi burada bir MusicianPlus class oluşturuyoruz ve Musician içindeki içeriklere erişmesi için inheritliyoruz.
    
    def __init__(self,name):#sadece self yazıp geçemezsin initializerıda aynı yazmak zorundasın.
        Musician.__init__(self,name)#burada da yine refer veriyoruz yukarıya.
        print("musician plus")#class tanımlanınca bu mesaj onayını alıyoruz.
    
    def test3(self):#bu da Musician classında olmayan ama Plus'ta olan bir method
        print("test3")#çalışınca bu yazsın istiyoruz.
                      #genel konsept olarak aslında musicianplus'ın içine musicianı atıp yeni methodlar yani buraya atadığımız insanlar
                      #eskiden olan sistemleride kullansın ama bu özelliklerede sahip olsun
                      #member,admin,manager,ceo gibi sınıflar oluşturduk gibi düşünebilirz yani erişim sağlayabilecek perkleri bir nevi oluşturuyoruz.
    #override yaptık test1 yerine koyduk ama ilk test1den Musician üstünden gidersek yine test1 çıktısını alırız.
    def test1(self):
        print("test1 test1")
```

```python
atlas = MusicianPlus("Atlas")
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.name = "Atlas Samancioglu"
```

```python
type(atlas)
```

## polymorphism

```python
class Banana():
    def __init__(self,name):
        self.name = name
        
    def info(self): #Muz sınıfının kapsamında olduğu için muz'u alır
        return f"100 calories {self.name}" #Muz 100 kalori gibi bir bilgi döndürüyor
class Apple():
    def __init__(self,name):
        self.name = name
    
    def info(self): #Elma sınıfının kapsamında olduğu için elmayı alır
        return f"150 calories {self.name}" #Elma 150 kalori gibi bir bilgi döndürür
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana,apple] #Muz ve Elma sınıfından iki ayrı obje ile liste yaptık
```

```python
for fruit in fruitList:
    print(fruit.info())
```

```python
#Farklı işlevler doğrultusunda aynı isimli metodları çalıştırabilmeye polymorphism denir
```

## encapsulation

```python
#Bir sınıfın ya da bir fonksiyonun içerisinde erişmemesi gereken bir kod varsa
#saklı tutmamıza yarar
```

```python
class Phone():

    def __init__(self,name,price):
        self.name = name
        self.__price = price #sadece sınıf içerisinden erişilip değiştirilebilir durumda
    
    def info(self):
        print(f"{self.name} price is: {self.__price}")

    def changePrice(self,price):
        self.__price= price
```

```python
iphone = Phone("Iphone 17",500)
```

```python
iphone.info()
```

```python
iphone.__price= 400 #fiyatını değiştirdik
```

```python
iphone.info()
```

```python
#bu fiyatın değiştirilmesini istemiyoruz diyelim
```

```python
#__price yaparsak eğer yani alt çizgi koyarsak sınıfın içerisinden erişilebilir yaparız.
```

```python
iphone.changePrice(300)
```

```python
iphone.info()
```

## abstraction

```python
#soyut bir sınıf oluşturuyoruz ama o sınıfın yapısını kullanarak ilerlememizi sağlıyor.
```

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):

    @abstractmethod
    def maxSpeed(self):
        pass
```

```python
#myCar = Car() Can't instantiate abstract class Car without an implementation der
```

```python
class Tesla(Car):
    
    def maxSpeed(self):
        print("200km")
```

```python
tesla = Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Mercedes(Car):

    def maxSpeed(self):
        print("250km")
```

```python
mercedes = Mercedes()
```

```python
mercedes.maxSpeed()
```

## special methods

```python
class Fruit():

    def __init__(self,name,calories):
        self.name = name
        self.calories = calories
    
    def __str__(self):
        return f"{self.name}: {self.calories} calories"
    
    def __len__(self):
        return self.calories
```

```python
myFruit = Fruit("banana", 150)
```

```python
myFruit.calories
```

```python
myFruit.name
```

```python
print(myFruit)
```

```python
len(myFruit)
```

```python
class Train():

    def __init__(self,name):
        self.name = name
    
    def __getitem__(self,key):
        if key == "a":
            return self.name
        else:
            return "Not found"
```

```python
myTrain = Train("myTrain")
```

```python
myTrain["a"]
```
