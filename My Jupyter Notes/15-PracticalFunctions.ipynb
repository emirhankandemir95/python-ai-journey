```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
my_list = [3,5,7,10,20,30]
```

```python
new_list = []
for i in my_list:
    result = divideNumber(i)
    new_list.append(result)
print(new_list)
```

## map

```python
help(map)
```

```python
list(map(divideNumber,my_list))
```

```python
def controlString(string):
    return "Emir" in string
```

```python
controlString("Emir Kandemir")
```

```python
controlString("Kandemir")
```

```python
myStringList = ["Emir","Emir Kandemir","Kandemir","Atlas"]
```

```python
list(map(controlString,myStringList))
```

## filter

```python
filter(controlString,myStringList)
```

```python
list(filter(controlString,myStringList)) #true olanları filterlayıp bize bir çıktı ver dedik
```

## lambda

```python
multiplyLambda = lambda num : num * 3 #def keyword kullanmadan fonksiyon oluşturduk böyle basit işlemleri tek satırda yazabilmek için kullanılır.
```

```python
type(multiplyLambda)
```

```python
multiplyLambda(20)
```

```python
result = multiplyLambda(10)
```

```python
result #return yazmışız gibi geri döndürüyor.
```
