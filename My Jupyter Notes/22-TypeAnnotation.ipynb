```python
def add_numbers(a: int,b: int) -> float: #float yazmamın sebebi sonucu değiştirmediğini göstermek amacıyla.
    return a + b                         #çünkü sadece kod okunabilirliğini artırmak amacıyla kullanılan
                                         #gösterim şekilleridir. böyle yazmanıza gerek yok.
```

```python
add_numbers(4,5)
```

```python
#python 3.10'dan azsa çalışmaz
def process_value(value: int | str) -> str:
    if isinstance(value,int):
        return f"processed integer: {value}"
    else:
        return f"processed string: {value}"
```

```python
process_value(50)
```

```python
process_value("emir")
```

```python
process_value([10,20,30])
```

```python
#python 3.9
from typing import List
def sum_list(numbers: List[int]) -> int:
    return sum(numbers)
```

```python
numbers = [1,2,3,4]
total = sum_list(numbers)
print(total)
```
