```python
class Person(): #class genelde büyük harfle başlar gibi düşün. karıştırmamak amacıyla.
    #property
    # name = ""
    # age = 0
    # gender = ""
    job = "" #Zorunlu değil. Çünkü aşağıda initializerda zorunluları yaptım zaten
    #propertyleri vermesende aşağıda bu özellikleri ben her türlü anlarım diyor
    #yorum satırı yaptığımda da aşağıdaki çıktılar değişmedi.
    #initializer method
    def __init__(self,name,age,gender): #Hata alırız çünkü eksik bir parametre var içerisine self koymak zorundayız.
        self.name = name
        self.age = age
        self.gender = gender
        # print("init executed") #init çalıştırıldı burayı yorum satırı yaptım çünkü aşağıda initialize ettiğimizde ne yazdığını göstermek için yazmıştım
    
    #method
    def printName(self):
        print(self.name)
```

```python
emir = Person("Emir",30,"Male")
```

```python
class Dog():

    year = 7

    def __init__(self,age=5):
        self.age = age
        self.dogHumanAge = age * self.year #illa initializer içinde yazılmamasına rağmen bu şekilde de bir tanım yapabiliriz.
        print("dog instance")
    def dogYearsInHuman(self):
        return self.age * Dog.year #Dog.year = > self.year
```

```python
myDog = Dog(3)
```

```python
myDog.age
```

```python
myDog.dogYearsInHuman()
```

```python
myDog.dogHumanAge #çağırdık mesela 21 sonucunu yine aldık
```
