```python
def args_sum(*args):
    return sum(args)
```

```python
args_sum(10,20,30,40,50,60)
```

```python
def args_example(*args):
    print(args)
```

```python
args_example(123,1203123,343,123,34)
```

```python
type(args_example)
```

```python
def kwargs_example(**kwargs):
    print(kwargs)
```

```python
kwargs_example(apple=100,banana=150,melon=200) #dictionary oluşturuyor ama fonksiyon
```

```python
type(kwargs_example)
```

```python
def kwargs_example_2(**kwargs):
    if "apple" in kwargs:
        print("appleeee")
    else:
        print(":(")
```

```python
kwargs_example_2(apple=10,banana=15)
```

```python
kwargs_example_2(banana=15)
```
