```python
while True:
    try:
         myAge = int(input("enter age: "))
         print(myAge * 2)
         #break
    except:
        print("enter your age!!!")
    else:
        print("else executed")
        break
    finally: #try - except var mı yok mu bakmaz hep çalışır.
        print("finally executed")
```
