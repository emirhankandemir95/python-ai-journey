```python
#6) Aşağıdaki string dizisinde, içinde sadece XYZ geçen barkodları gösterecek yeni bir liste oluşturunuz
```

```python
barkodDizisi = ["ABC231","SA3123XYZ","XYZA123Q","QRE1231KJ","X112QGL"]
```

```python
# cevap
def controlString(s):
    return "XYZ" in s

new_list = list(filter(controlString,barkodDizisi))
print(new_list)
```

```python
#8) Aşağıda yazdırılan sınıfı incelediğinizde kedim.yasiCarp() kodunun çıktısı ne olacaktır?
```

```python
class Kedi():
        
    def __init__(self,isim,yas=5):
        self.isim = isim
        self.yas = yas
        
    def yasiCarp(self):
        return self.yas * 3
```

```python
kedim = Kedi("Tonton")
```

```python
# kedim.yasiCarp()
# yaş değeri 5 belirtildiği için aşağıda değer verilmedi için 15 olur
```

```python
#9) Aşağıdaki kodun çıktısı ne olacaktır?
```

```python
class Ogrenci():
    
    def __init__(self,isim,sinavNotu):
        self.isim = isim
        self.__sinavNotu = sinavNotu
    
    def notuGoster(self):
        print(f"{self.isim} sınav notu: {self.__sinavNotu}")
```

```python
ogrenci = Ogrenci("Mehmet",85)
```

```python
ogrenci.__sinavNotu = 75
```

```python
# ogrenci.notuGoster()
# yine 85 çıkar çünkü "__" ile kilitlemişiz. Dışarıdan erişim sağlayamayız.
```
