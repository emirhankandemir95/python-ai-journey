```python
def divideNumber(number):
    return number / 2
```

```python
divideNumber(20)
```

```python
my_list = [3,5,7,10,20,30]
```

```python
new_list = []
for i in my_list:
    result = divideNumber(i)
    new_list.append(result)
print(new_list)
```
