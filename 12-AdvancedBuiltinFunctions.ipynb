## range function

```python
range(50) #0'dan 50'ye kadar sayı dizimi
```

```python
list(range(10)) #10'a kadar bir liste oluştur
```

```python
for num in list(range(20)):#20 ye kadar oluşturulan bir listenin her elemanlarına num de
    print(num * 2)#sonra bu num değerini 2 ile çarp
```

```python
list(range(5,25)) #range(starting index,stopping index,stepping size) virgül ile ama
```

```python
list(range(5,25,2)) #5'ten 25'e kadar yap 2'şer 2'şer atla
```

## diyelim ki bu sayıları değilde indexlerini yazdırmak istiyorum

## 1. yol

```python
for ix in range(len(my_list)):#ix yani index gibi düşün len(my_list) yani uzunluğunda bir range oluşturduk
    print(ix)#indexi yazdırdık
```

```python
for ix in range(len(my_list)):
    print(my_list[ix])
```
